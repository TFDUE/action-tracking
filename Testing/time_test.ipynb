```python
import time

def ackermann(m, n):
    if m == 0:
        return n + 1
    elif n == 0:
        return ackermann(m - 1, 1)
    else:
        return ackermann(m - 1, ackermann(m, n - 1))

timings = []

for i in range(100):
    start = time.time()
    
    ackermann(3,8)
    
    end = time.time()
    
    timings.append(end-start)
    
print(timings)
print("max:" + str(max(timings)))
print("min:" + str(min(timings)))
print("mean:" + str(sum(timings)/len(timings)))
```
